The percentage change in fuel cost per unit burned in 1998 compared to 1994 for fuel type coal

```python
coal_fuel_cost_per_unitburn = raw_df[['fuel_cost_per_unit_burned','report_year','fuel_type_code_pudl']][raw_df['fuel_type_code_pudl'] == 'coal']
coal_fuel_cost_per_unitburn
```

```python
diff_94 = coal_fuel_cost_per_unitburn['fuel_cost_per_unit_burned'][coal_fuel_cost_per_unitburn['report_year']== 1994].max() - coal_fuel_cost_per_unitburn['fuel_cost_per_unit_burned'][coal_fuel_cost_per_unitburn['report_year']== 1994].min()
```

```python
diff_98 = coal_fuel_cost_per_unitburn['fuel_cost_per_unit_burned'][coal_fuel_cost_per_unitburn['report_year']== 1998].max() - coal_fuel_cost_per_unitburn['fuel_cost_per_unit_burned'][coal_fuel_cost_per_unitburn['report_year']== 1998].min()
```

```python
diff_94
```
