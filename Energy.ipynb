```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
```

```python
import pandas as pd
```

```python
new = pd.read_csv(url)
new
```

```python
new.describe(include = 'all')
```

```python
new.drop(['date', 'lights'], axis='columns', inplace=True)
new
```

```python
y = new['Appliances']
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
new_df = pd.DataFrame(scaler.fit_transform(new), columns = new.columns)
```

```python
X = new_df.drop('Appliances', axis='columns')
X
```

importing necessary library

```python
from sklearn.metrics import  mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
import numpy as np
```

```python
def get_weights_df(model, feat, col_name):
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two D.P?

```python
x2 = np.array( X.T2).reshape((-1, 1))
y2 = np.array(X.T6)
model2 = LinearRegression()
model2.fit(x2,y2)
```

```python
pred2 = model2.predict(x2)
```

```python
r2_score = r2_score(y2, pred2)
round(r2_score, 2)
```

What is the Mean Absolute Error (in two decimal places)?

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, pred)
round(mae, 2)
```

What is the Residual Sum of Squares (in two decimal places)?

```python
rss = np.sum(np.square(y_test - pred))
round(rss, 3)
```

What is the Root Mean Squared Error (in three decimal places)?

```python
rmse = np.sqrt(mean_squared_error(y_test, pred))
round(rmse, 3)
```

What is the Coefficient of Determination (in two decimal places)? it is the same as r2_score 

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred)
round(r2_score, 2)
```

Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively? 

```python
linear_model_weights = get_weights_df(model, x_train, 'Linear_Model_Weight')
linear_model_weights

```

Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights? ans 26

```python
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
lasso_weights_df
```

What is the new RMSE with the Lasso Regression (in 3 decimal places)?

```python
laspred = lasso_reg.predict(x_test)
```

```python
rmsel = np.sqrt(mean_squared_error(y_test, laspred))
round(rmsel, 3)
```
